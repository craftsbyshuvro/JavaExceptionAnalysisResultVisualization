```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
from prettytable import PrettyTable
import re
from pandasql import sqldf
import tldextract
import requests
import validators
import glob

BASE_DATA_PATH = 'D:/Study/Graduate Seminar/Data/'
file = open(BASE_DATA_PATH+'FinallyBlockMethods.txt', "r")
InvokedMethods= file.read()
InvokedMethods = InvokedMethods.split(',')
totalInvokedMethods= len(InvokedMethods)
topInvokedMethods = pd.DataFrame({'freq': InvokedMethods})
fig, ax = plt.subplots(figsize=(10,7))
import brewer2mpl
bmap = brewer2mpl.get_map('Set2','qualitative',8,reverse=True)
colors = bmap.mpl_colors
ax.set_title("TFrequently Called Methods in Finally Block")
ax.set_xlabel('Frequency')
ax.set_ylabel('Invoked Methods')
width = 0.75
alltopInvokedMethods = topInvokedMethods.freq.value_counts().nlargest(10).sort_values(ascending=True)
ind = np.arange(len(alltopInvokedMethods))
ax.barh(alltopInvokedMethods.index,alltopInvokedMethods, width, color=colors)
for i, v in enumerate(alltopInvokedMethods):
    ax.text(v + 5, i - .15, str(v),color = 'black')
    ax.text(v + 5, i + .25, '('+str(round(v/totalInvokedMethods*100))+'%)',color = 'blue')
plt.show()
```
